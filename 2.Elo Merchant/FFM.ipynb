```python
train.replace([np.inf, -np.inf], np.nan,inplace=True)
train.fillna(0,inplace = True)
```

```python
train.shape
```

```python
def numeric_to_cat(df,col_name):
#     cat_name = [col_name+'_'+str(x) for x in range(4)]
#     print(cat_name)
    df = pd.qcut(df,[0, .50, .75, 1.],
#                  labels=cat_name,
                 duplicates='drop'
                )
    return df

```

```python
excluded =['card_id']
col_to_cat = [col for col in train.columns if col not in excluded]
for col in col_to_cat:
    train[col] = numeric_to_cat(train[col],col)
```
